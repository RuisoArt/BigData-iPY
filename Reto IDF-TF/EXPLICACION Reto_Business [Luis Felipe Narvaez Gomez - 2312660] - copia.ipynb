## Metodo SK-Learn

Dentro del trabajo “IDF-TF Modificado” observamos una cierta diferencia entre los resultados arrojados entre el código escrito en base a la teoría vista sobre TF e IDF contra la función de SK-Learn. En este trabajo prefiero utilizar SK-Learn por la severidad de los resultados y gusto personal al utilizar un entrenamiento con una red neuronal.

```python
def SKL_TF_IDF(parrafo):
    tr_idf_model  = TfidfVectorizer()
    tf_idf_vector = tr_idf_model.fit_transform(parrafo)

    tf_idf_array = tf_idf_vector.toarray()
    
    words_set = tr_idf_model.get_feature_names()
    
    df_tf_idf_generic = pd.DataFrame(tf_idf_array, columns = words_set)
    
    return df_tf_idf_generic
```

## Metodo de Limpieza de texto

El siguiente método se encarga de quitar de nuestro texto todos aquellos caracteres, números, símbolos, emojis, letras especiales o cualquier cosa innecesaria en el texto o que pueda confundir a nuestra máquina. Aquí debemos tener en cuenta que una vez leamos el archivo el mismo se presentara como un único STRING, una línea de texto continua. Aquí aprovecharemos esto y le podremos decir a nuestra maquina que busque entre la línea de STRING los caracteres que no necesitamos y los elimine.

Para esto utilizamos la función REPLACE, la cual escanea un String buscando un carácter dado y reemplazándolo por otro.


```python
# Este metodo esta basado en el taller anterior que utiliza la funcion replace para limpiar el texto
def clean_data(paragraph):
    my_paragraph = paragraph
    # letras en mayusculas a minusculas
    my_paragraph = my_paragraph.lower()
    # primero las vocales con tilde o cosas raras que no deberia tener una vocal en ingles
    my_paragraph = my_paragraph.replace('á','a')
    my_paragraph = my_paragraph.replace('é','e')
    my_paragraph = my_paragraph.replace('í','i')
    my_paragraph = my_paragraph.replace('ó','o')
    my_paragraph = my_paragraph.replace('ú','u')
    my_paragraph = my_paragraph.replace('â','a')
    # los caracteres
    my_paragraph = my_paragraph.replace(',','')
    my_paragraph = my_paragraph.replace(';','')
    my_paragraph = my_paragraph.replace(':','')
    my_paragraph = my_paragraph.replace('.','')
    my_paragraph = my_paragraph.replace('¨','')
    my_paragraph = my_paragraph.replace('´','')
    my_paragraph = my_paragraph.replace('{','')
    my_paragraph = my_paragraph.replace('}','')
    my_paragraph = my_paragraph.replace('[','')
    my_paragraph = my_paragraph.replace(']','')
    my_paragraph = my_paragraph.replace('^','')
    my_paragraph = my_paragraph.replace('`','')
    my_paragraph = my_paragraph.replace('*','')
    my_paragraph = my_paragraph.replace('~','')
    my_paragraph = my_paragraph.replace('+','')
    my_paragraph = my_paragraph.replace('?','')
    my_paragraph = my_paragraph.replace('str(\)','')
    my_paragraph = my_paragraph.replace('¿','')
    my_paragraph = my_paragraph.replace('¡','')
    my_paragraph = my_paragraph.replace('|','')
    my_paragraph = my_paragraph.replace('°','')
    my_paragraph = my_paragraph.replace('¬','')
    my_paragraph = my_paragraph.replace('!','')
    my_paragraph = my_paragraph.replace('"','')
    my_paragraph = my_paragraph.replace('#','')
    my_paragraph = my_paragraph.replace('$','')
    my_paragraph = my_paragraph.replace('%','')
    my_paragraph = my_paragraph.replace('&','')
    my_paragraph = my_paragraph.replace('/','')
    my_paragraph = my_paragraph.replace('(','')
    my_paragraph = my_paragraph.replace(')','')
    my_paragraph = my_paragraph.replace('=','')
    my_paragraph = my_paragraph.replace('@','')
    # Eliminar numeros
    my_paragraph = my_paragraph.replace('1','')
    my_paragraph = my_paragraph.replace('2','')
    my_paragraph = my_paragraph.replace('3','')
    my_paragraph = my_paragraph.replace('4','')
    my_paragraph = my_paragraph.replace('5','')
    my_paragraph = my_paragraph.replace('6','')
    my_paragraph = my_paragraph.replace('7','')
    my_paragraph = my_paragraph.replace('8','')
    my_paragraph = my_paragraph.replace('9','')
    my_paragraph = my_paragraph.replace('0','')
    # Espacios y otros caracteres mas extraños
    my_paragraph = my_paragraph.replace('“','')
    my_paragraph = my_paragraph.replace('  ','')
    my_paragraph = my_paragraph.replace('”','')
    my_paragraph = my_paragraph.replace('\n\n','')
    my_paragraph = my_paragraph.replace('\n','')
    #simbolos de dinero
    my_paragraph = my_paragraph.replace('£','')
    my_paragraph = my_paragraph.replace('€','')
    my_paragraph = my_paragraph.replace('¥','')
    my_paragraph = my_paragraph.replace('元','')
    my_paragraph = my_paragraph.replace('₩','')
    my_paragraph = my_paragraph.replace('₹','')
    my_paragraph = my_paragraph.replace('₽','')
    my_paragraph = my_paragraph.replace('₺','')
    my_paragraph = my_paragraph.replace('ł','')
    my_paragraph = my_paragraph.replace('č','')
    my_paragraph = my_paragraph.replace('₪','')
    my_paragraph = my_paragraph.replace('₱','')
    
    return my_paragraph
```

El método funciona de manera sencilla, por el entra por parámetro nuestro STRING (la noticia) y de manera sucesiva va reemplazando cada carácter que nosotros digamos que quite por nada (eliminándolo), tambien trabaja desde el principio llevando el texto a solo minusculas para evitar trabajar con las mismas palabras pero entendidas doble vez por la maquina, por último nos retorna un texto limpio.

## Limpiar palabras

El siguiente método tiene la función de quitar ahora palabras no deseadas.

Al momento de clasificar noticias y ver las palabras mas relevantes que describan este texto, tenemos el problema que la maquina no distinguirá una palabra con significado de una que solo sirve como contexto o conector en el lenguaje. Aquí podemos hacer dos cosas, o tener un algoritmo que pueda distinguir el contexto de un texto y quitarme perfectamente las palabras contextuales pero carentes de significado argumental, o … simplemente quitar esas palabras del texto para que no las analice. Tomare la ruta dos.

Aquí el problema radica en la robustes de nuestro diccionario de palabras que vamos a necesitar. El diccionario en primera será diferente dependiendo de idioma que estemos trabajando, no es lo mismo los conectores y palabras de contexto, así como indicativos, pronombres y otros; propios de un idioma que de otro. En segunda se debe tener cuidado con que palabras se quieren quitar, puesto que alguna como wish=desear, puede contextual e irrelevante en un texto dado, pero en otro relevante para la clasificación. En tercera, debido a que no se puede hacer un diccionario de palabras lo suficientemente robusto y acertado para cada noticia, indudablemente se colaran algunas palabras que pueden llegar a ser tomadas como etiquetas, en este caso, dependerá del ingeniero analizar las mismas para la clasificación con otros filtros automatizados o a criterio personal.

Así pues, al siguiente método ingresa un arreglo de diferentes STRINGs separados por comas. El mismo arreglo será recorrido por un ciclo FOR para eliminar la palabra completa que vaya encontrando de acuerdo al diccionario que nosotros preparamos.

De aquí saltamos a un problema y es que las palabras eliminadas dejan dentro del arreglo su espacio lleno con un String vacío, para esto se recorre nuevamente la lista de palabras eliminando cada espacio ocupado por un string vacío.

Por último, nos retorna una lista de palabras ya limpio.


```python
# Ahora limpiar palabras que no se necesitan

def clean_works(paragraph):
    my_paragraph = paragraph

    longitud_data = int(len(my_paragraph))
    json = ['the','from','for','in','one','bn','aol','said','which','set','was','also','our','sec','than','add','ad','all','way','had','with','on','as','loss','win','hot','has','m','are','will','could','am','into','but','now','its','when','i','he','she','it','you','yours','they','we','lower','upper''had','have','can','cant','that','this','yes','no','what','why','how','many','at','us','to','is','a','an','of','and','or','by','were','where','up','down','right','left']
    longitud_json = int(len(json))
    
    # quita palabras con el diccionario
    buscar = json
    sustituir = ''
    for i in range(longitud_data):
        for j in range(longitud_json):
            if my_paragraph[i] == buscar[j]:
                my_paragraph[i] = sustituir
    
    #crea un arreglo nuevo para quitar los elementos vacios
    my_paragraph_2 =[]
    for val in my_paragraph: 
        if val != '' : 
            my_paragraph_2.append(val) 
                
    return my_paragraph_2
```

```python


def read_archives(ruta):
    corpus = []
    direccion_archivo = ruta
    # Lectura y añadir al corpus
    my_file = open(direccion_archivo, mode = "r")
    businness_001 = (my_file.read())
    my_file.close()
    businness_001 = clean_data(businness_001)
    businness_001 = businness_001.split(' ')
    businness_001 = clean_works(businness_001)
    corpus = corpus + businness_001 # seria bueno para sacar etiquetas por todo el lote
    #corpus.append(businness_001) # seria bueno para sacar etiquetas por cada archivo

    #Crear al dataframe
    df_business = SKL_TF_IDF(corpus)

    #crear nueva fila con la sumatoria de las columnas
    nueva_fila = df_business.sum()
    df_business = df_business.append(nueva_fila, ignore_index=True)

    #ordenar de manera decendente(mayor a menor)
    df_business_2 = df_business.copy()
    df_business_2 = df_business_2.sort_index(ascending=False)

    # Eliminar las otras filas
    #numero de filas
    num_rows = int(df_business.shape[0]) - 1
    df_business_2 = df_business_2.drop(range(0,num_rows),axis=0)

    #Hacer la transpuesta baby
    df_money = df_business_2.transpose()
    #Ordenar la transpuesta
    #nombre primera columna
    name_first_column = int(df_money.columns.values[0])
    df_money = df_money.rename(columns={name_first_column:'Valor'})
    df_money = df_money.sort_values('Valor',ascending=False)

    #obtener las PRIMERAS 10 PALABRAS
    valores = df_money.head(10).index.values

    #Escribir en los archivos las etiquetas
    my_file_two = open(direccion_archivo,"a")
    my_file_two.write("\n"+"keywords: {}".format(valores))
    my_file_two.close()
    
    return print("revisa los archivos")
```

Le quedaría mejor un nombre de … Main a esta sección del notebook. Lo siguiente no es un método, es nuestro “menú” en el cual haremos uso de lo anteriormente programado para empezar a recoger y grabar cada noticia.

Aquí el problema que se resuelve en el código es tratar de leer cada una de las diferentes noticias que contine un Folder, para ello hay que idear una manera de recorrer la carpeta con un ciclo programado como si de un arreglo se tratara. Aquí aprovechamos que en esta situación concreta las noticias poseen como nombre una numeración consecutiva y finita. Por lo que aprovechando la variable de iteración [i] de un ciclo FOR, podemos recorrer las noticias ordenadamente, la única traba es que la numeración empieza desde 001 a XXX no de 1 a X, por lo que solo podemos utilizar la iteración en un numero limitado de nombres dentro de las rutas del archivo. Por suerte podemos hacer tres recorridos FOR para esto de la siguiente manera: [001 – 009] [010 – 099] [100 – XXX] siendo los números las variables del rango de iteración de cada uno de los tres FOR respectivos para cada recorrido.

En síntesis, para leer un archivo se necesita de una ruta y el nombre del archivo, el nombre del mismo es un numero que se puede iterar de forma ordenada, se hacen tres recorridos por cada tipo de nombre estableciendo rangos específicos. El número máximo será el ingresado por el tercer rango del tercer FOR interpretado por el tamaño total de archivos por folder + 1. Cada iteración por dentro enviará la dirección al método que leerá la noticia y dará como resultado las etiquetas o palabras claves, este a su vez se apoya de los métodos de limpieza para mejorar su hallazgo.

Y de esta manera ya tenemos cada noticia con sus keywords
