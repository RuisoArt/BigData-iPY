### Random Forest

Para la reinterpretación de este algoritmo tuvimos que tener en cuenta el cómo funciona internamente el mismo y no solo tomar este como una caja negra que usar en nuestros problemas; de hecho la selección de los diferentes algoritmos aquí utilizados pueden interpretarse mejor al saber como funcionan en realidad por dentro.

Al recurrir a este algoritmo tenemos que tener en cuenta que el "padre”, si así lo pudiéramos llamar, son el conjunto de algoritmos de **Forest** en lo que encontramos **Decision Tree** y el que al final utilizamos que es **Random Forest**.

El **árbol de decisiones** es la forma que sencilla de clasificación por preguntas de decisión que podemos encontrar, cada uno de los nodos partiendo desde la raíz hasta la última de las ramas esta divida en la **interpretación binomial** de la respuesta de una pregunta (SI se cumple, No se cumple) de esta manera podemos **clasificar grupos de datos** que cumplan con una serie de características ejemplificadas en nodos de las ramas. 

Para crear el árbol de decisiones se requiere de un **dataset** con el que vamos a trabajar, en este caso se presupone por la naturaleza del algoritmo que ya se tiene una etiqueta (algoritmo de aprendizaje supervisado).

Lo que hace internamente Random Forest es **comparar cada una de las columnas, renglón a renglón, con nuestra etiqueta**, con la finalidad de lograr predecir que rasgos de nuestros datos funcionan mejor para cada etiqueta. **¿El problema?** los datasets normalmente pueden llegar a tener impurezas, es decir, columnas que en sus filas tengas **datos nulos o dañados** y puedan alterar la decisión del algoritmo en fijarse que rasgos son claves para definir una etiqueta. **¿La solución?** Hay algoritmos de respaldo que revisan la impureza que pueda tener un dato, como por ejemplo **GINI IMPURITY**.

Ahora bien **Los bosques Aleatorios** funcionan basados en los bosques de decisión, de hecho, para hacer un random forest, se necesitan crear decisions Trees; sin embargo, trabaja un tanto diferente.

**¿Cuál es la diferencia?** 
1. primero es que no se toma el **dataset completo**, de hecho, se toman **100 datos aleatorios** del **dataset** original y se crea uno **totalmente nuevo**. No se puede saber que datos son y hasta pueden haber repetidos.
2. El nuevo dataset se le llama **Bootstrapped Dataset** En base al se crea un Decision Tree, la diferencia es que, el decisión Tree normal, revisa fila a fila, columna por columna el dataset, cogiendo cada rasgo con su etiqueta; aquí NO, se **toman rasgos aleatorios**, unas columnas si, otras no, en base a esos rasgos **se crean el Desicion Tree**.
3. Todo esto se hace un **numero "n" de veces**.

Al final tenemos un numero de "n" arboles construidos por Decision Trees creados por la selección RANDOM de datos del dataset original y con rasgos/columnas RANDOM que estimar con su etiqueta.

Así, cada árbol tendrá su propia predicción al momento de clasificar un dato en determinada etiqueta, en conjunto dan un voto (como en una democracia) y esto da un score de clasificación.

Es decir, la suma de veredictos de clasificación, al ser diferentes formas de estimar una etiqueta, nos entrega un puntaje de predicción más alto y confiable que con un solo árbol.

### Support Vector Machine

Las máquinas de soportes de vectores, son una excelente opción a elegir al momento de querer clasificar una serie de datos dependiendo de su etiqueta.

Este algoritmo nativamente es utilizado para el aprendizaje supervisado, s decir que los datos tienen una serie de etiquetas que ayudan con el aprendizaje del mismo algoritmo, sin embargo, también existe su versión para aprendizaje no supervisado ampliamente utilizado en trabajos de Deep Learning, cosa que no veremos en este momento.

La forma de trabajar de SVM o las máquinas de soportes de vectores es, como su nombre lo dice, a través del uso de uno o varios vectores que nos ayuden en el momento de separar uno o más grupos en un diagrama, y es que SVM expresa a manera de grafico 2D (x,y) cada uno de los datos como un punto coordenado en dicho plano.

Según la semejanza de estos datos en sus rasgos o datos por columnas que posean, estarán en un lado u otro de la gráfica 2D, luego de la impresionen el grafico de cientos de datos daremos en que muchos de estos datos pueden visualizarse como agrupaciones y una simple forma de separarlos tanto visual como matemáticamente en el plano es mediante el uso de una recta.



Para determinar el vector que separa uno o más grupos se utiliza una tarea que se encargue de evaluar la distancia que hay entre los grupos de datos, estimando una margen igual por ambos lados de la recta, asegurando la homogeneidad de la margen, esto nos ayuda a delimitar el área en la que se puede colocar un dato en el plano y el mismo empieza a ser parte de un grupo, ayudando a su clasificación.



Los problemas empiezan cuando tenemos muchos grupos de datos o muchas categorías donde a su vez los datos pueden verse mezclados entre grupos y sui discriminación es más complicada, por lo que su locion radica en alterar la forma del vector a partir de comportamientos geométricos, matemáticos y pliométricos.

"variando el parametro C"


Esta última solución también se puede ampliar con lo que conocemos como los kernel de SVM, que no son más que distintas formas de realización del vector que nos limitan el área donde puede estar o no un dato en el plano 2D para su discriminación.

### K-Means

K-Means ... Este algoritmo no nos revelo mucho durante el experimento de uso, de hecho, no nos arrojó nada a la predicción de datos, la respuesta de esto está en cómo funciona por dentro.

K-means o K-medias en español, es un algoritmo de aprendizaje no supervisado, es decir, que no necesita etiquetas para poder agrupar datos o clasificarlos, el solo se encarga de ver los rasgos de determinados datos y según la similitud de estos juntarlos en bajo una sola etiqueta que los defina.

La manera de hacer esto es imprimiendo en un gráfico 2D cada uno de los diferentes datos de nuestro data set, cada uno con una coordenada (x,Y) dada por las caracteristicas o rasgos intrínsecos al dato. Luego de esto K-means se encarga de plantear un numero n de clusters o centroides claves al azar utilizando datos cualesquiera del gráfico, a partir de ellos evalúa que datos se encuentran más cercanos de determinado cluster y así arma n grupos con los mismo n cluster, cada grupo es una nueva etiqueta.

El proceso se repite n veces sea necesaria hasta que se empiece a hallar un patrón de agrupamiento repetitivo y satisfactorio en el que podemos decir que los datos están bien clasificados.

De esta manera el algoritmo K-means prueba:

1. Es un algoritmo rápido, robusto y simple que proporciona resultados confiables cuando los conjuntos de datos son distintos o bien separados entre sí de forma lineal.
2. Debemos especificar el número de clusteres, de antemano y los resultados finales son sensibles a la inicialización.
3. La agrupación puede no funcionar bien si contiene datos muy superpuestos o si los datos son ruidosos o están llenos de valores atípicos.

Dicho esto **¿Cuál fue el problema con Kmeans?** resulta que nuestro dataset en primera si posee etiquetas, de hecho maneja dos columnas de rasgos como ellas, esto de por si no es un problema para forzar la implementación de k-means, de hecho se podría utilizar esta información para mejorar la colocación de cada dato en un plano 2D, sin embargo lo que obtenemos es ruido que se vuelve difícil de trabajar para K-means.

En segunda, los rasgos de cada fila de texto son limitados y nada claros, en vez de ser datos que sirvan para una clara identificación en (X,Y) para el plano 2D, son opiniones que terminan por generar una distribución de puntos sin agrupación aparente, con datos solapándose y con bastante ruido.

Puesto que los datos están muy muy juntos entre sí y hay mucho ruido como para determinar realmente conjuntos de datos, a pesar de ser un algoritmo rápido y robusto, el solo cálculo de numero de clusters puede llevar hasta 22h sin siquiera sacar un resultado propio.

En el caso de OpinionesVSatracciones solo deberíamos tomar 3 clusters, uno por cada etiqueta posible, así mismo en OpinionesVSpolaridades deberían ser de solo 5, sin embargo hallando el k de cluster necesarios para clasificación, la predicción no era ni cerca de ser satisfactoriamente positiva ni con un K=1000, esto debido al propio ruido de datos existente y el problema de determinación de distancias entre tantos puntos.
