Este conjunto de datos está tomado de Serial Killers por número de víctimas (Wikipedia) y contiene cuatro archivos .csv.
Highest_victim_count.csv contiene el nombre, el país, los años activos, las víctimas comprobadas, las posibles víctimas y el resumen de los delitos cometidos por los asesinos en serie con el mayor número de víctimas.

15_to_30_victim_count.csv contiene los mismos detalles para los asesinos en serie con un recuento de 15 a 30 víctimas.

5_to_14_victim_count.csv contiene los mismos detalles para los asesinos en serie con un recuento de 5 a 14 víctimas.

### Datasets Concatenados

La mejor forma de trabajar con este extenso dataframe es concatener cada una de la partes separadas, de la (1) a la (4) y trabajarlos como uno solo.

```python
df = pd.concat(map(pd.read_csv, [datasets_number_one,
                                 datasets_number_two,
                                 datasets_number_three,
                                 datasets_number_four]), ignore_index=True)
```
